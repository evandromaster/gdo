```python
import pandas as pd
import numpy as np
from datetime import datetime
df = pd.read_sql_table('tbl_envolvidos', 'sqlite:///gdo.db',
                       columns =['NREDS', 'COD_DIAO', 'T_C', 'COD_DELITO', 'T_C_DELITO','GRAU_LESAO','GRUPO_TIPO_ENVOL',
                                 'DATA_FATO', 'CIA', 'UEOP', 'QTDE_ENVOL'])
```
