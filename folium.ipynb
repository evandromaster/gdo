```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import pendulum
df = pd.read_sql_table('tbl_crimes', 'sqlite:///gdo.db',
                       columns =['NREDS', 'COD_DIAO', 'COD_DIAO1', 'COD_DIAO2','COD_DIAO3','T_C', 'DATA_FATO', 'MUNICIPIO',
                                 'LATITUDE','LONGITUDE','CIA', 'UEOP', 'QTDE'])
df
```

```python
import folium
from folium import plugins
mapa = folium.Map(
                  location=[-20.145554,-44.890541], #coordenada  de divinópolis
                  tiles= 'OpenStreetMap',    #'Stamen Toner' #'Stamen Terrain'
                  width = '100%',
                  heigth = '100%',
                  zoom_start=8
                 )  
mapa.add_child(folium.LatLngPopup())#Exibindo as coordenadas de Latitude e Longitude no evento de click
mapa
```

```python
# coordenadas=[]
# for LATITUDE,LONGITUDE in zip(df.LATITUDE.values[:18000],df.LONGITUDE.values[:18000]):
#     coordenadas.append([LATITUDE,LONGITUDE])

coordenadas = df[['LATITUDE', 'LONGITUDE', 'QTDE']]
coordenadas= coordenadas.dropna()
coordenadas
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))        
mapa
```

```python
mapa.save('mapa-consumidor.html')
```

```python
for i in range(0, len(coordenadas)):
    folium.Circle(
        location = [coordenadas.iloc[i]['LATITUDE'], coordenadas.iloc[i]['LONGITUDE']],
        color = '#00FF69',
        fill = '#00A1B3',
        toltip = '<li><bold> CIDADE: ' + str(df.iloc[i]['MUNICIPIO']),
        #radius = (coordenadas.iloc[i]['QTDE']**1.1)
        radius = 0
    ).add_to(mapa)
mapa
```
